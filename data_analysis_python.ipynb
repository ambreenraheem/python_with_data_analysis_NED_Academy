#### Preprocessing of DataFrame

#### Basic data cleaning.
##### Step 1: Convert "Price" to numeric, replacing, '$' , and converting 'Free' to '0'.
##### " $ " is Special Character so need to replace it from Price.

```python
mydf['Price'] = mydf['Price'].replace('Free' , '0')
```

```python
mydf['Type'].unique()
```

```python
type(mydf['Price'])
```

```python
mydf['Price'] = mydf['Price'].str.replace('$','').astype(float)
```

```python
mydf['Price'].unique()
```

```python
mydf.info()
```

```python
mydf.head()
```

#### Additional Steps
#### Step 2: Remove duplicates based on the 'App' column
If there is any duplicate values in perticular column, it will drop it or you can say remove it from dataframe.

```python
mydf = mydf.drop_duplicates(subset=['APP'])
```

```python
mydf.head()
```

#### Step 3: Fill missing values in 'Rating' with the mean rating. If there is any missing value in your dataframe.

```python
mydf['Rating'] = mydf['Rating'].fillna(mydf['Rating'].mean())
```

```python
mydf.head()
```

#### Step 4: Normalize the 'Review' column (scaling between 0 and 1)

```python
mydf['Reviews'] = (mydf['Reviews'] - mydf['Review'].min() / (mydf['Reviews'].max() - mydf['Reviews'].min))
```

#### Step 5: Simplify 'Size' column (replace 'Varies with device' with NaN )

```python
mydf['Size'] = mydf['Size'].replace('Varies with device', np.nan)
```

```python
mydf.head()
```

#### Step 6: Cnvert 'Size' from string to numeric (assume 'M' means MB), beacouse it is a special character.

```python
mydf['Size'] = mydf['Size'].str.replace('M','').astype(float)
```

```python
mydf.head()
```

#### Step 7: Fill missing values in 'Size' with the median size...Agr kisi ko median ki base pr fillna krna chahte hyn to.

```python
mydf['Size'] = mydf['Size'].fillna(mydf['Size'].median())
```

```python
mydf.head()
```

Now it's NaN values replaced on median based..

#### Step 8: Clean 'Installs' column (remove '+' and ','), because + and , are special characters.

```python
mydf['Installs'] = mydf['Installs'].replace(',','').replace('+','').astype(int)
```

```python
mydf.head()
```

#### Step 11: Create a new column 'Popularity' based on 'Rating' and 'Reviews'

```python
mydf['Popularity'] = mydf['Rating'] * mydf['Reviews']     # mere data m se rating or reviews k columns k base pr ek naya column popularity bnao
```

```python
mydf.sort_values(by=['Popularity','Size'], ascending=False) # yhan hm ab popularity or size k sath sortout kro
```

```python
mydf['Affordable'] = np.where(mydf['Type'] == 'Paid' , np.where(mydf['Price'] < 5.00, 'Yes', 'No'), 'N/A')
```
