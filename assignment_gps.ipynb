## Preprocessing of the DataFrame
### Basic data cleaning.
#### "We used one-hot encoding during the preprocessing of the DataFrame to convert categorical variables into numerical format."
#### "Before analyzing the data, we performed preprocessing on the DataFrame to handle missing values and outliers."

```python
googleplay_store['Rating'] = googleplay_store['Rating'].fillna(0.0)
```

```python
googleplay_store['Rating'].astype(float)
```

```python
googleplay_store['Rating'].nunique()
```

```python
googleplay_store['Type']= googleplay_store['Type'].replace('Free','nan').replace('nan',0)
```

```python
googleplay_store['Type']= googleplay_store['Type'].fillna(0)
```

```python
googleplay_store['Type'].unique()
```

```python
googleplay_store['Type'].info()
```

```python
googleplay_store['Price']= googleplay_store['Price'].astype(float)
```

```python
googleplay_store['Price'].info()
```

```python
googleplay_store.info()
```

```python
googleplay_store['Size'] = (
    googleplay_store['Size']
    .str.replace('M', '000000', regex=False)
    .str.replace('K', '000', regex=False)
    .str.replace('Varies with device', '0', regex=False)
    .str.replace('.', '', regex=False)
    .replace('1,000+', '1000')
)
```

```python
googleplay_store['Size']= googleplay_store['Size'].str.strip("k ").astype(float)
```

```python
googleplay_store['Size'].astype(float)
```

```python
googleplay_store['Size'].info()
```

```python
googleplay_store.info()
```

```python
googleplay_store.head()
```

#### Step 2: Remove duplicates based on the 'App' column
If there is any duplicate values in perticular column, it will drop it or you can say remove it from dataframe.

```python
googleplay_store.drop_duplicates(subset=['App'])
```

```python
googleplay_store['Installs'] = (
    googleplay_store['Installs']
    .str.replace('+', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace(',', '', regex=False)
    .replace('0', '',regex=False)
    .str.replace('Free','00')
)
```
